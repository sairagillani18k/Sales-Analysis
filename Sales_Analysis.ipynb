# Business Q-1: Best month in terms of Sales?

```python
sales_each_month = all_data.groupby('Month')["Sales","Quantity Ordered","Price Each"].sum()
sales_each_month
```

```python
fig,ax = plt.subplots(figsize=(14,6))
no_of_months = ["January","February","March","April","May","June","July","August","September","October","November","December"]
ax.set_title('Trend of Sales per Month',fontsize=14)
ax.set_xlabel('Month',fontsize=13)
ax.set_ylabel('Sales ($)',fontsize=13)
bars = ax.bar(no_of_months,sales_each_month["Sales"],0.35,color='g')
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        # '%g' % is done to drop trainling zeros from decimal position
        ax.annotate('%g' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color='r', fontsize=10, rotation=90)


autolabel(bars)
```

From the bar graph we can see "December" is best month in terms of Sales.

# Business Q-4: What products are most often sold together?

```python
# try to find duplicate order Id's to get the products sold together
duplicate_df = all_data[all_data['Order ID'].duplicated(keep=False)]
print(duplicate_df.shape)
duplicate_df.head(10)
```

```python
duplicate_df['Groupped_Products'] = duplicate_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
duplicate_df = duplicate_df[['Order ID', 'Groupped_Products']].drop_duplicates()
#This line of code will drop the duplicates and will just store the Order ID and Groupped_Products in the duplicate_df

print(duplicate_df.shape)
duplicate_df.head()
```

```python
# Now we need to count the pairs which occur together. We need to use some other libraries for this task

from itertools import combinations
from collections import Counter

count = Counter()

for row in duplicate_df['Groupped_Products']:
    row_list = row.split(',')           #converting the row values to list with comma split
    count.update(Counter(combinations(row_list, 2)))
                 
                 
# print(count)
# count.most_common(10)
#most_common methods only brings the first 10 items.


for k, v in count.most_common(10):
    print(k, v)
```

```python
# if we need most commonly sold 3 items in a row

count = Counter()

for row in duplicate_df['Groupped_Products']:
    row_list = row.split(',')           #converting the row values to list with comma split
    count.update(Counter(combinations(row_list, 3)))
    
for k, v in count.most_common(10):
    print(k, v)
    
# and if you need more combination, keep increasing the second argument of .update(Counter(Combinations(row_list, ?)))
```

# Business Quesiton 5: What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = all_data.groupby('Product')['Quantity Ordered'].sum()


products = [product for product, df in product_group]   #fetching individual products from the product_group


# simple plotting
fig = plt.figure(figsize=(14,5))
plt.bar(products, quantity_ordered, color='g')
plt.title("Trend of sales of each Product over last 1 year", fontsize=15)
plt.xticks(products, rotation="vertical")
plt.ylabel('Quantity Ordered', fontsize=13)
plt.xlabel('Product', fontsize=13)
plt.show()
```

Looking at the barGraph it can be seen that (AAA Batteries and AA Batteries) are the top sold items. But reason is not clear from the barGraph.

```python
price = all_data.groupby('Product').mean()['Price Each']
price
```

```python
#Let's plot both the sales mean and product sold most


#creating subplot for two plotting in same area.
fig, ax1 = plt.subplots(1,1,figsize=(14,5))

#twinning the axes
ax2 = ax1.twinx()

# 1st plot on 1st axis
ax1.bar(products,quantity_ordered, ec='black', label="Qauntity Ordered", color='g')

#2nd plot on second axis in same area as axis 1
ax2.plot(products,price, color="b", label="Price ($)")

# applying legends to both plots/axis
ax1.legend(loc=0)
ax2.legend(loc="upper left")

#adding title to the chart
plt.title("Products vs. Quantity Ordered vs. Unit Price")

#adding axis labels to both axis.
ax1.set_xlabel('Product Name', size=14)
ax1.set_ylabel('Quantity Ordered', size=14)
ax1.set_xticklabels(products, rotation='vertical', size=12)
ax1.set_ylim(0,35000)
ax2.set_ylabel('Price($)', size=14)



#adding number of quantity sold on each bar.
for i in range(len(quantity_ordered)):
    ax1.text(i,quantity_ordered[i], quantity_ordered[i], ha='center', va='bottom',
             rotation='horizontal', fontweight='semibold', color='r')
```

Looking at above graph we can conclude that Batteries are the most sold items because their prices are low and are more frequently bought items because used in many household items (almost all elctronic devices, wallclock, baby toys, etc.), also Batteries are not long-lasting. On the other hand items like LG Dryer and Washing Machine are expensive and long lasting items and monstly one item is used per houshold, so these items are sold minimum no.of times.
